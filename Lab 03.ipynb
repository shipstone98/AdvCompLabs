#### Module Imports

```python
from matplotlib import pyplot
import numpy, pandas
```

## Section 01 - Fit

```python
x = numpy.array([-3, 2, -1, 4, 0])
y = numpy.array([20, 20, -10, 30, 0])
cubic_eqn = numpy.polyfit(x, y, 3)
print(cubic_eqn)
```

```python
xp = numpy.linspace(-4, 6, 100)
poly_values = numpy.polyval(cubic_eqn, xp)
pyplot.title("Graphics of data points and fitted cubic equation")
pyplot.plot(x, y, '.')
pyplot.plot(xp, poly_values)
pyplot.show()
```
