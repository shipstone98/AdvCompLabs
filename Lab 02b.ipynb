## Exercise 03
#### Constants

```python
PRED_POP = 15
PREY_POP = 110

PRED_BIRTHS = 0.02
PREY_BIRTHS = 2.2

PRED_DEATHS = 1.08
PREY_DEATHS = 0.04
```

#### Module Imports

```python
from matplotlib import pyplot
```

### Parts A and B

```python
def model_pred_prey(time_step, sim_len):
    pred_pop = PRED_POP
    prey_pop = PREY_POP
    t = 0

    t_list = [ t ]
    pred_list = [ pred_pop ]
    prey_list = [ prey_pop ]

    while t <= sim_len:
        t += time_step
        pred_births = pred_pop * PRED_BIRTHS * prey_pop
        prey_births = prey_pop * PREY_BIRTHS
        pred_deaths = pred_pop * PRED_DEATHS
        prey_deaths = prey_pop * PREY_DEATHS * pred_pop
        pred_pop += time_step * (pred_births - pred_deaths)
        prey_pop += time_step * (prey_births - prey_deaths)
        t_list.append(t)
        pred_list.append(pred_pop)
        prey_list.append(prey_pop)

    return t_list, pred_list, prey_list
```

### Part C

```python
t_list, pred_list, prey_list = model_pred_prey(0.001, 15)
pyplot.title("Predator/prey simulation")
pyplot.ylabel("Population")
pyplot.xlabel("Time (hrs)")
pyplot.plot(t_list, pred_list, label="Predators")
pyplot.plot(t_list, prey_list, label="Prey")
pyplot.legend()
pyplot.show()
```
